```python
import numpy as np
import pandas as pd
from scipy.stats import norm, truncnorm
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# define the features to consider
features = ["size", "weight", "smoothness", "red", "green", "blue"]

# number of features and samples
n_f = len(features)
n_s = 256
```

# Generate features

## Apple features

```python
# generate the apple sizes
mu = 7 # (cm)
sigma = 1.5 # (cm)
apple_size = norm(mu, sigma).rvs(n_s) # (cm)

plt.hist(apple_size);
```

```python
# generate apple weights
mu = 85 # (g)
sigma = 10 # (g)
apple_weight = norm(mu, sigma).rvs(n_s) # (cm)

plt.hist(apple_weight);
```

```python
# generate apple smoothness
mu = 9.5 # (a.u.)
sigma = 0.2 # (a.u.)
min_val = 0 # (a.u.)
max_val = 10 # (a.u.)
apple_smoothness = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s) # (a.u.)

plt.hist(apple_smoothness);
```

```python
# generate apple red value
mu = 230 # (uint8)
sigma = 30 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
apple_red = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
apple_red = np.uint8(np.round(apple_red)) # (uint8)

plt.hist(apple_red);
```

```python
# generate apple green value
mu = 15 # (uint8)
sigma = 5 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
apple_green = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
apple_green = np.uint8(np.round(apple_green)) # (uint8)

plt.hist(apple_green);
```

```python
# generate apple blue value
mu = 0 # (uint8)
sigma = 3 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
apple_blue = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
apple_blue = np.uint8(np.round(apple_blue)) # (uint8)

plt.hist(apple_blue);
```

```python
# get an impression of the orange colors
apple_color = np.stack((apple_red, 
                        apple_green, 
                        apple_blue), 
                       axis=1)
plt.scatter(np.random.rand(n_s), np.random.rand(n_s), s=20, c=apple_color / 255)
```

```python
# convert apple features to dataframe
apple = {
    "size": apple_size,
    "weight": apple_weight,
    "smoothness": apple_smoothness,
    "red": apple_red,
    "green": apple_green,
    "blue": apple_blue,
}
apple = pd.DataFrame(apple)

apple.head()
```

## Orange features

```python
# generate the orange sizes
mu = 7 # (cm)
sigma = 2 # (cm)
orange_size = norm(mu, sigma).rvs(n_s) # (cm)

plt.hist(orange_size);
```

```python
# generate orange smoothness
mu = 5 # (a.u.)
sigma = 0.66 # (a.u.)
min_val = 0 # (a.u.)
max_val = 10 # (a.u.)
orange_smoothness = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s) # (a.u.)

plt.hist(orange_smoothness);
```

```python
# generate orange red value
mu = 235 # (uint8)
sigma = 10 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
orange_red = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
orange_red = np.uint8(np.round(orange_red)) # (uint8)

plt.hist(orange_red);
```

```python
# generate orange green value
mu = 140 # (uint8)
sigma = 10 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
orange_green = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
orange_green = np.uint8(np.round(orange_green)) # (uint8)

plt.hist(orange_green);
```

```python
# generate orange blue value
mu = 0 # (uint8)
sigma = 3 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
orange_blue = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
orange_blue = np.uint8(np.round(orange_blue)) # (uint8)

plt.hist(orange_blue);
```

```python
# get an impression of the orange colors
orange_color = np.stack((orange_red, 
                        orange_green, 
                        orange_blue), 
                       axis=1)
plt.scatter(np.random.rand(n_s), np.random.rand(n_s), s=20, c=orange_color / 255)
```

```python
# convert orange features to dataframe
orange = {
    "size": orange_size,
    "weight": orange_weight,
    "smoothness": orange_smoothness,
    "red": orange_red,
    "green": orange_green,
    "blue": orange_blue,
}
orange = pd.DataFrame(orange)

orange.head()
```

```python
# add labels to the data
apple["label"] = "apple"
orange["label"] = "orange"
# merge both dataframes
fruits = pd.concat((apple, orange), ignore_index=True)
# shuffle the rows and remove old indices
fruits = fruits.sample(frac=1).reset_index(drop=True)

fruits.tail()
```

# Machine Learning

## Preprocessing

```python
# split the data in training and testing
train, test = train_test_split(fruits, test_size=0.2, random_state=42)

# drop the labels
train_label = train.pop("label")
test_label = test.pop("label")
# convert the labels to 0s and 1s
label_encoder = LabelEncoder()
label_encoder.fit(["orange", "apple"])
train_label = label_encoder.transform(train_label)
test_label = label_encoder.transform(test_label)

# scale the training data
min_max_scaler = MinMaxScaler()
train_scaled = min_max_scaler.fit_transform(train) 
test_scaled = min_max_scaler.transform(test)
```

## Unsupervised learning

```python
# initialize the PCA model
pca = PCA(n_components=n_f)

# fit the PCA model to the training data
train_pca = pca.fit_transform(train_scaled)
# get the explained variance ratios
exp_var_n2 = np.sum(pca.explained_variance_ratio_[:2])

print(f"The first to principle components explain {exp_var_n2 * 100:.1f}% of the variance in the data")
```

```python
# show the weight of the features in the components
loadings = pd.DataFrame(pca.components_[:2].T, columns=['PC1', 'PC2'], index=train.columns)
loadings
```

```python
plt.scatter(train_pca[:, 0], train_pca[:, 1], facecolor=(0, 0, 0), s=6)
```

```python
# define maximum number of clusters
n_cluster_max = 8
n_cluster = np.arange(1, n_cluster_max + 1)
scores = np.empty(n_cluster_max)

for ii in range(n_cluster_max):
    # create model
    k_means = KMeans(n_cluster[ii])
    # fit the model to the data
    k_means.fit(train_pca[:, :2])
    # score the model
    scores[ii] = k_means.score(train_pca[:, :2])

# plot the scores as a function of clusters
plt.plot(n_cluster, scores)
```

```python
# fit the model to the training data
k_means = KMeans(2)
k_means.fit(train_pca[:, :2])

# get the k-means
cluster_centers = k_means.cluster_centers_
```

```python
# predict the training labels
train_label_predict = k_means.predict(train_pca[:, :2])

# plot the transformed train data
train_pca_0 = train_pca[train_label_predict==0, :]
train_pca_1 = train_pca[train_label_predict==1, :]

plt.scatter(train_pca_0[:, 0], train_pca_0[:, 1], facecolor=(0.3, 0.7, 0.9), s=6)
plt.scatter(train_pca_1[:, 0], train_pca_1[:, 1], facecolor=(1, 0.4, 0.2), s=6)
plt.scatter(cluster_centers[0, 0], cluster_centers[0, 1], facecolor=(0, 0, 0), marker="x", s=100)
plt.scatter(cluster_centers[1, 0], cluster_centers[1, 1], facecolor=(0, 0, 0), marker="x", s=100)
```

```python
# transform the test data
test_pca = pca.transform(test_scaled)

# predict the labels of the test data
test_label_predict = k_means.predict(test_pca[:, :2])

# plot the transformed train data
test_pca_0 = test_pca[test_label_predict==0, :]
test_pca_1 = test_pca[test_label_predict==1, :]

plt.scatter(test_pca_0[:, 0], test_pca_0[:, 1], facecolor=(0.3, 0.7, 0.9), s=6)
plt.scatter(test_pca_1[:, 0], test_pca_1[:, 1], facecolor=(1, 0.4, 0.2), s=6)
plt.scatter(cluster_centers[0, 0], cluster_centers[0, 1], facecolor=(0, 0, 0), marker="x", s=100)
plt.scatter(cluster_centers[1, 0], cluster_centers[1, 1], facecolor=(0, 0, 0), marker="x", s=100)
```
