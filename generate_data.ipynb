```python
from os.path import join
import numpy as np
import pandas as pd
from scipy.stats import norm, truncnorm
import matplotlib.pyplot as plt
```

```python
# define the features to consider
features = ["size", "weight", "smoothness", "red", "green", "blue"]

# number of features and samples
n_f = len(features)
n_s = 512
```

# Generate features

## Apple features

```python
# generate the apple sizes
mu = 7 # (cm)
sigma = 1.5 # (cm)
apple_size = norm(mu, sigma).rvs(n_s) # (cm)

plt.hist(apple_size);
```

```python
# generate apple weights
mu = 85 # (g)
sigma = 10 # (g)
apple_weight = norm(mu, sigma).rvs(n_s) # (cm)

plt.hist(apple_weight);
```

```python
# generate apple smoothness
mu = 9.5 # (a.u.)
sigma = 0.2 # (a.u.)
min_val = 1 # (a.u.)
max_val = 10 # (a.u.)
apple_smoothness = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s) # (a.u.)

plt.hist(apple_smoothness);
```

```python
# generate apple red value
mu = 210 # (uint8)
sigma = 30 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
apple_red = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
apple_red = np.uint8(np.round(apple_red)) # (uint8)

plt.hist(apple_red);
```

```python
# generate apple green value
mu = 30 # (uint8)
sigma = 10 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
apple_green = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
apple_green = np.uint8(np.round(apple_green)) # (uint8)

plt.hist(apple_green);
```

```python
# generate apple blue value
mu = 0 # (uint8)
sigma = 5 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
apple_blue = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
apple_blue = np.uint8(np.round(apple_blue)) # (uint8)

plt.hist(apple_blue);
```

```python
# get an impression of the orange colors
apple_color = np.stack((apple_red, 
                        apple_green, 
                        apple_blue), 
                       axis=1)
plt.scatter(np.random.rand(n_s), np.random.rand(n_s), s=20, c=apple_color / 255)
```

```python
# convert apple features to dataframe
apple = {
    "size": np.round(apple_size, 1),
    "weight": np.round(apple_weight, 1),
    "smoothness": np.round(apple_smoothness, 1),
    "red": apple_red,
    "green": apple_green,
    "blue": apple_blue,
}
apple = pd.DataFrame(apple)

apple.head()
```

## Orange features

```python
# generate the orange sizes
mu = 7 # (cm)
sigma = 2 # (cm)
orange_size = norm(mu, sigma).rvs(n_s) # (cm)

plt.hist(orange_size);
```

```python
# generate orange weights
mu = 140 # (g)
sigma = 20 # (g)
orange_weight = norm(mu, sigma).rvs(n_s) # (cm)

plt.hist(orange_weight);
```

```python
# generate orange smoothness
mu = 7.5 # (a.u.)
sigma = 1 # (a.u.)
min_val = 1 # (a.u.)
max_val = 10 # (a.u.)
orange_smoothness = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s) # (a.u.)

plt.hist(orange_smoothness);
```

```python
# generate orange red value
mu = 235 # (uint8)
sigma = 5 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
orange_red = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
orange_red = np.uint8(np.round(orange_red)) # (uint8)

plt.hist(orange_red);
```

```python
# generate orange green value
mu = 110 # (uint8)
sigma = 20 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
orange_green = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
orange_green = np.uint8(np.round(orange_green)) # (uint8)

plt.hist(orange_green);
```

```python
# generate orange blue value
mu = 0 # (uint8)
sigma = 5 # (uint8)
min_val = 0 # (uint8)
max_val = 255 # (uint8)
orange_blue = truncnorm(loc=mu, scale=sigma, a=(min_val - mu) / sigma, b=(max_val - mu) / sigma).rvs(n_s)
orange_blue = np.uint8(np.round(orange_blue)) # (uint8)

plt.hist(orange_blue);
```

```python
# get an impression of the orange colors
orange_color = np.stack((orange_red, 
                        orange_green, 
                        orange_blue), 
                       axis=1)
plt.scatter(np.random.rand(n_s), np.random.rand(n_s), s=20, c=orange_color / 255)
```

```python
# convert orange features to dataframe
orange = {
    "size": np.round(orange_size, 1),
    "weight": np.round(orange_weight, 1),
    "smoothness": np.round(orange_smoothness, 1),
    "red": orange_red,
    "green": orange_green,
    "blue": orange_blue,
}
orange = pd.DataFrame(orange)

orange.head()
```

```python
# add labels to the data
apple["label"] = "apple"
orange["label"] = "orange"
# merge both dataframes
fruits = pd.concat((apple, orange), ignore_index=True)
# shuffle the rows and remove old indices
fruits = fruits.sample(frac=1).reset_index(drop=True)

fruits.tail()
```

```python
# define the filepath
filepath = join("data", "fruits.csv")

# save the dataframe
fruits.to_csv(filepath, index=False)
```
